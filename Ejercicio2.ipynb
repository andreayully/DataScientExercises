#### Analisis de la columna "marital", la cual indica si esta casado 

```python
dataset.marital.value_counts()

```

```python
dataset.marital.unique()
```

```python
dataset.marital.isnull().sum()
```

```python
columna_estado_civil= dataset.marital.apply(
    lambda estado_civil: estado_civil.strip() if type(estado_civil)== str else estado_civil)
```

```python
columna_estado_civil.value_counts()
```

```python
#Reemplazar la columna marital con la procesada
copia_dataset = dataset.copy()
copia_dataset[["marital"]] = columna_estado_civil
```

```python
copia_dataset.marital.value_counts()
dataset.marital.value_counts()
```

```python
copia_dataset.isnull().sum()
```

```python
#Analisis de la columna "Education"
```

```python
copia_dataset.loc[copia_dataset.education.isnull(), "education"] = "unknown" 
```

```python
#Copia de la copia del dataset original
copia_dataset_default = copia_dataset.copy()
```

```python
copia_dataset_default.default.replace("no", "No", inplace=True)
copia_dataset_default.default.replace("nO", "No", inplace=True)
copia_dataset_default.default.replace("N", "No", inplace=True)
copia_dataset_default.default.replace("yes", "Yes", inplace=True)
copia_dataset_default.default.replace("YES", "Yes", inplace=True)
```

```python
copia_dataset_default.default.value_counts()
```

```python
copia_dataset_default.isnull().sum()
```

```python
datos_columnas_analizadas = copia_dataset_default.loc[:, ["default", "education", "marital", "job", "no.hijos", "child", "edad"]]
```
